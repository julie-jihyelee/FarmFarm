```python
from datetime import datetime, timedelta

# 전날 날짜 계산
yesterday = datetime.now() - timedelta(days=1)
yesterday_formatted = yesterday.strftime('%Y-%m-%d')

# 전날 날짜에서 년, 월, 일 분리
year = yesterday.year
month = yesterday.month
day = yesterday.day

print('전날 날짜:', yesterday_formatted)
print('년:', year)
print('월:', month)
print('일:', day)
```
