#### 전처리

```python
# 가격 날짜 2023년으로 되어있어 해당년도 년으로 변경필요
# 가격 4년단위로 날짜 2/29인 이상 데이터 있는것들 제거필요. (2012년, 2016년)
# 날씨 시작부분에 있는 기간별조회 라는 필요없는 데이터 삭제
# 가격,날씨 확장자가 xls 이므로 xlsx 로 변환 필요


import numpy as np
import pandas as pd
import glob
import datetime as dt


def Prepro(name, start_year, end_year):
    
    # 가격 엑셀 불러오기
    df_prices = []  # 불러온 엑셀 파일들을 저장할 리스트

    for number in range(start_year, end_year + 1):
        file_name = f"./{name}/{name}가격{number}.xlsx"
        df_price = pd.read_excel(file_name, index_col='구분')
        df_prices.append(df_price)
        
    # 가격 행열 전치
    for i in range((end_year - start_year)+1):
        df_prices[i] = df_prices[i].T
    
    # 가격 연도 전부 합치기
    df_price_total = pd.concat(df_prices)
    
    df_price = df_price_total[['평균']]
    
    
    # ,제거 함수
    def remove_comma(x):
        if isinstance(x, str):
            return int(x.replace(',', ''))
        else:
            return x
    
    # ,제거
    df_price['평균'] = df_price['평균'].apply(remove_comma)


    # 증감치 컬럼 추가
#     df_price['증감치'] = df_price['평균'].diff()  
    
    
    
    
    # 날씨 엑셀 불러오기
    df_weathers = []

    for number in range(start_year, end_year+1):
        file_name = f"./{name}/{name}날씨{number}.xlsx"
        df_weather = pd.read_excel(file_name, index_col = '날짜')
        df_weathers.append(df_weather)
    
    # 날씨 처음 4개 행 제거
    for i in range((end_year - start_year)+1):
        df_weathers[i] = df_weathers[i].iloc[5:]
        
    # 날씨 연도 전부 합치기
    df_weather_total = pd.concat(df_weathers)
    df_weather = df_weather_total[['최고기온','최저기온','강수량']]
    
    # 강수량 - => 0
    df_weather.replace("-", 0, inplace=True)
    # 강수량 int 타입으로 변경
    df_weather['강수량'].astype(int)
    
    # 날씨 index(날짜)의 datetype이 달라 머신러닝을 위한 datetime으로 변경
    df_weather.index = pd.to_datetime(df_weather.index)
 
    df_weather = df_weather.drop('강수량', axis=1)
    
    # 가격에는 빈 날짜가 존재하여 날씨를 기준으로 합치기
    df = pd.concat([df_weather, df_price], axis=1)
    
    df = df.rename(columns={'평균':'가격'})
    
    df.reset_index(inplace=True)
    df.rename(columns={'index': '날짜'}, inplace=True)
    
    # 날짜를 월, 일, 연도로 분리하여 새로운 컬럼으로 추가
    df['월'] = df['날짜'].dt.month
    df['일'] = df['날짜'].dt.day
    df['연도'] = df['날짜'].dt.year    
    
    df.set_index('날짜', inplace=True)
   
    # 증감치 컬럼 추가
#     df['증감치'] = df['가격'].diff()
   
    df['가격2'] = df['가격']
    df = df.drop('가격', axis=1)
    df.rename(columns={'가격2': '가격'}, inplace=True)


    # Nan값 drop
    df = df.dropna()
    
    df.to_csv(f'{name}.csv', index_label='날짜')
    
    return df
```

```python
Prepro('홍고추', 2011, 2022)
```

#### 머신러닝

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor

def machine(name):
    # csv 불러오기
    df = pd.read_csv(f'{name}.csv', index_col='날짜')

    X = df.loc[:, '최고기온':'연도']
    y = df['가격']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                  test_size=0.25)
    
    m1 = DecisionTreeRegressor()
    m2 = RandomForestRegressor()
    m3 = ExtraTreesRegressor()
    
    m1.fit(X_train, y_train)
    m2.fit(X_train, y_train)
    m3.fit(X_train, y_train)
    
    print(m1.score(X_test, y_test), m2.score(X_test, y_test), m3.score(X_test, y_test))
    
    return df
```

```python
machine('감자')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor

def machine_p(name, HT, LT, month, day, year):
    # csv 불러오기
    df = pd.read_csv(f'{name}.csv', index_col='날짜')

    X = df.loc[:, '최고기온':'연도']
    y = df['가격']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                  test_size=0.25)
    
    m1 = DecisionTreeRegressor()
    m2 = RandomForestRegressor()
    m3 = ExtraTreesRegressor()
    
    m1.fit(X_train, y_train)
    m2.fit(X_train, y_train)
    m3.fit(X_train, y_train)
    
    print(m1.score(X_test, y_test), m2.score(X_test, y_test), m3.score(X_test, y_test))
    
    print(m1.predict([[HT, LT, month, day, year]]))
    print(m2.predict([[HT, LT, month, day, year]]))
    print(m3.predict([[HT, LT, month, day, year]]))
```

#### R2, mse, rmse

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.metrics import mean_squared_error, r2_score

def machine(name):
    # csv 불러오기
    df = pd.read_csv(f'{name}.csv', index_col='날짜')

    X = df.loc[:, '최고기온':'연도']
    y = df['가격']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                  test_size=0.25)
    
    m1 = DecisionTreeRegressor()
    m2 = RandomForestRegressor()
    m3 = ExtraTreesRegressor()
    
    m1.fit(X_train, y_train)
    m2.fit(X_train, y_train)
    m3.fit(X_train, y_train)
    
    y_pred_m1 = m1.predict(X_test)
    y_pred_m2 = m2.predict(X_test)
    y_pred_m3 = m3.predict(X_test)
    
    mse_m1 = mean_squared_error(y_test, y_pred_m1)
    mse_m2 = mean_squared_error(y_test, y_pred_m2)
    mse_m3 = mean_squared_error(y_test, y_pred_m3)
    
    r2_m1 = r2_score(y_test, y_pred_m1)
    r2_m2 = r2_score(y_test, y_pred_m2)
    r2_m3 = r2_score(y_test, y_pred_m3)

    rmse_m1 = np.sqrt(mse_m1)
    rmse_m2 = np.sqrt(mse_m2)
    rmse_m3 = np.sqrt(mse_m3)
    
    print(f"Decision Tree R2 Score: {r2_m1}")
    print(f"Random Forest R2 Score: {r2_m2}")
    print(f"Extra Trees R2 Score: {r2_m3}")
    
    print(f"Decision Tree MSE: {mse_m1}")
    print(f"Random Forest MSE: {mse_m2}")
    print(f"Extra Trees MSE: {mse_m3}")
    
    print(f"Decision Tree RMSE: {rmse_m1}")
    print(f"Random Forest RMSE: {rmse_m2}")
    print(f"Extra Trees RMSE: {rmse_m3}")
    
    return df
```

```python
machine('홍고추')
```
